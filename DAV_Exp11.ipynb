```r
import matplotlib.pyplot as plt

marks=[98, 87, 91, 56, 73]
names=["Soham", "Arpit", "Parth", "Sahil", "Ajay"]
plt.bar(names, marks, width=0.5)
plt.title("Students v/s Marks", fontsize=25, color="green")
plt.ylabel("Marks out of 100")
plt.xlabel("Names of Students")
plt.show()
```

```r
library(ggplot2)
```

```r
# Sample data
x <- c(1, 2, 3, 4, 5)
y <- c(2, 4, 5, 4, 3)

# Basic scatter plot
ggplot(data.frame(x, y), aes(x = x, y = y)) + geom_point()
```

```r
# Sample data
x <- c(1, 2, 3, 4, 5)
y <- c(1, 3, 5, 4, 2)

# Line plot
ggplot(data.frame(x, y), aes(x = x, y = y)) + geom_line()
```

```r
library(lattice)
```

```r
x <- c(1, 2, 3, 4, 5)
y <- c(2, 4, 5, 4, 3)

# Scatter plot
xyplot(y ~ x)
```

```r
categories <- c("A", "B", "C")
values <- c(10, 20, 15)

# Bar chart
barplot(values ~ categories)
```
